```python
from sklearn.preprocessing import StandardScaler as sc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Housing.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(labels=["hotwaterheating"],axis=1,inplace=True)
```

```python
scaler=sc()
```

```python
# scaler.fit(df.loc[:,['area', 'bedrooms', 'bathrooms', 'stories', 'parking']])
```

```python
val= scaler.fit_transform(df.loc[:,['area', 'bedrooms', 'bathrooms', 'stories', 'parking']])
```

```python
(df.loc[:,['area', 'bedrooms', 'bathrooms', 'stories', 'parking']])=val
```

```python
data_scaled = df.copy()
data_encoded = pd.get_dummies(data_scaled, drop_first=True)
```

```python
data_encoded

```

```python
data_encoded.columns
```

```python
x=data_encoded.loc[:,[ 'area', 'bedrooms', 'bathrooms', 'stories', 'parking',
       'mainroad_yes', 'guestroom_yes', 'basement_yes', 'airconditioning_yes',
       'prefarea_yes', 'furnishingstatus_semi-furnished',
       'furnishingstatus_unfurnished']]
y=data_encoded.iloc[:,0]
```

```python
# data_encoded.drop(labels=["furnishingstatus_semi-furnished"],axis=1,inplace=True)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train,x_test,y_train,y_test=tts(x,y,test_size=.25)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression as lr
```

```python
model=lr()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import r2_score as r2
from sklearn.metrics import mean_squared_error as ms
from sklearn.metrics import mean_absolute_error as ma
```

```python
print(f"R2_score is {r2(y_test,y_predict)}")
print(f"Mean_squared_error is {ms(y_test,y_predict)}")
print(f"Mean_absolute_error is {ma(y_test,y_predict)}")
```
